```python
tc_harold_path = gpd.read_file("TC_Harold_path.geojson")
```

```python
tc_harold_path = tc_harold_path.to_crs(3832)
tc_harold_path.plot()
```

```python
tc_harold_path.crs
```

```python
tc_harold_buffer_10m = tc_harold_path.buffer(10000)
tc_harold_buffer_50m = tc_harold_path.buffer(50000)
tc_harold_buffer_100m = tc_harold_path.buffer(100000)
```

```python
tc_harold_buffer_10m.to_file("tc_harold_buffer_10km.geojson")
tc_harold_buffer_50m.to_file("tc_harold_buffer_50km.geojson")
tc_harold_buffer_100m.to_file("tc_harold_buffer_100km.geojson")
```
