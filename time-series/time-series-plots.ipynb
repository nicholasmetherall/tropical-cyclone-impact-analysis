```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap 
!pip install openpyxl
import openpyxl
```

```python
tcl_fiji = pd.read_csv("tree-cover-loss-fiji.csv")
tcl_vanuatu = pd.read_csv("tree-cover-loss-vanuatu.csv")
```

```python
tcl_fiji
```

```python
# Create a DataFrame
df = pd.DataFrame(tcl_fiji)

# Set 'Year' as the index
df.set_index('Year', inplace=True)

# Create a stacked bar plot
df.plot(kind='bar', stacked=True)
```

```python
# List of colors for each bar
colors = ['maroon', 'orange', 'yellow', 'grey']

# Plot with specific colors for each stack
ax = df.plot(kind='bar', stacked=True, y=['Western', 'Northern','Central', 'Eastern'], color=colors)

# Add titles and labels
ax.set_title('Tree cover loss by division in Fiji from 2000 to 2023')
ax.set_xlabel('Years')
ax.set_ylabel('Tree cover loss (square km)')

plt.show()
```

```python
tcl_vanuatu.rename(columns={'Unnamed: 0': 'Year'}, inplace=True)
tcl_vanuatu
```

```python
# Create a DataFrame
df_vu = pd.DataFrame(tcl_vanuatu)

# Set 'Year' as the index
df_vu.set_index('Year', inplace=True)

# Create a stacked bar plot
df_vu.plot(kind='bar', stacked=True)
```

```python
# List of colors for each bar
colors = ['black', 'maroon', 'red', 'orange', 'yellow', 'grey']

# Tafea	Shefa	Malampa	Penama	Sanma	Torba


# Plot with specific colors for each stack
ax = df_vu.plot(kind='bar', stacked=True, y=['Tafea', 'Shefa','Malampa', 'Penama', 'Sanma', 'Torba'], color=colors)

# Add titles and labels
ax.set_title('Tree cover loss by division in Vanuatu from 2000 to 2023')
ax.set_xlabel('Years')
ax.set_ylabel('Tree cover loss (square km)')

plt.show()
```
